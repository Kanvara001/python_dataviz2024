## หาคนที่เป็นมะเร็งต่อมน้ำเหลือง

```python
# prompt: แสดงคนที่เป็นมะเร็งโดยดูจาก comments ทั้งหมด

# Filter comments containing "มะเร็ง"
cancer_comments = Docter_df[Docter_df['comments'].str.contains('มะเร็ง', na=False)]
cancer_comments
```

```python
# prompt: แสดงคนที่เป็นมะเร็งต่อมน้ำเหลือง โดยดูจาก comments

# Filter comments containing "มะเร็งต่อมน้ำเหลือง"
lymphoma_comments = Docter_df[Docter_df['comments'].str.contains('มะเร็งต่อมน้ำเหลือง', na=False)]
lymphoma_comments
```

```python
# prompt: หาจำนวนคนที่เป็นมะเร็งต่อมน้ำเหลือง

# Calculate the number of people interested in lymphoma
num_lymphoma_people = len(lymphoma_comments)

print(f"Number of people interested in lymphoma: {num_lymphoma_people}")
```

```python
# prompt: คำนวณค่าเฉลี่ยจำนวนคนที่เป็นมะเร็งต่อมน้ำเหลืองจากคนที่เป็นมะเร็งทั้งหมด

# Calculate the number of people interested in cancer
num_cancer_people = len(cancer_comments)

# Calculate the proportion of people interested in lymphoma out of those interested in cancer
proportion_lymphoma = num_lymphoma_people / num_cancer_people

print(f"Proportion of people interested in lymphoma out of those interested in cancer: {proportion_lymphoma:.4f}")
```
