```python
#แสดงตารางข้อมูลจากไฟล์ CSV
import pandas as pd
from IPython.display import display

# อ่านไฟล์ CSV
person = pd.read_csv('/content/drive/My Drive/Midterm_BasicProgramming/Exam2/person.csv')
interview = pd.read_csv('/content/drive/My Drive/Midterm_BasicProgramming/Exam2/interview.csv')
income = pd.read_csv('/content/drive/My Drive/Midterm_BasicProgramming/Exam2/income.csv')
get_fit_now_member = pd.read_csv('/content/drive/My Drive/Midterm_BasicProgramming/Exam2/get_fit_now_member.csv')
get_fit_now_check_in = pd.read_csv('/content/drive/My Drive/Midterm_BasicProgramming/Exam2/get_fit_now_check_in.csv')
facebook_event_checkin = pd.read_csv('/content/drive/My Drive/Midterm_BasicProgramming/Exam2/facebook_event_checkin.csv')
drivers_license = pd.read_csv('/content/drive/My Drive/Midterm_BasicProgramming/Exam2/drivers_license.csv')
crime_scene_report = pd.read_csv('/content/drive/My Drive/Midterm_BasicProgramming/Exam2/crime_scene_report.csv')

# สร้างรายการ DataFrame
dfs = [person, interview, income, get_fit_now_member, get_fit_now_check_in, facebook_event_checkin, drivers_license, crime_scene_report]

# แสดงข้อมูลจากทุก DataFrame
for df in dfs:
    display(df)
```

```python
# prompt: หาคนที่ขับรถ Tesla Model S

# หาคนที่ขับรถ Tesla Model S
tesla_model_s_drivers = drivers_license[drivers_license['car_make'] == 'Tesla']
tesla_model_s_drivers = tesla_model_s_drivers[tesla_model_s_drivers['car_model'] == 'Model S']
display(tesla_model_s_drivers)
```

```python
# prompt: หาคนที่เข้าร่วม SQL Symphony Concert 3 ครั้งในเดือนธันวาคม 2017

# หาคนที่เข้าร่วม SQL Symphony Concert 3 ครั้งในเดือนธันวาคม 2017
december_concerts = facebook_event_checkin[(facebook_event_checkin['event_name'] == 'SQL Symphony Concert') & (facebook_event_checkin['date'].between(20171201, 20171231))]
concert_attendees = december_concerts.groupby('person_id').size()
frequent_attendees = concert_attendees[concert_attendees == 3].index.tolist()

# แสดงข้อมูลของคนที่เข้าร่วม 3 ครั้ง
SQL_Symphony_Concert_3times = person[person['id'].isin(frequent_attendees)]
display(SQL_Symphony_Concert_3times)
```

```python
# prompt: หาคนที่มีlicense_idจากSQL_Symphony_Concert_3times ที่ตรงกับ driver_license

# หาคนที่มี license_id จาก SQL_Symphony_Concert_3times ที่ตรงกับ driver_license
matching_drivers = drivers_license[drivers_license['id'].isin(SQL_Symphony_Concert_3times['license_id'])]

# แสดงข้อมูลของคนขับที่ตรงกัน
display(matching_drivers)
```

```python
# prompt: แสดงข้อมูลส่วนตัวของคนที่มี plate_number 500123

# หาคนที่มี plate_number 500123
person_with_plate = drivers_license[drivers_license['plate_number'] == '500123']

# ดึง license_id ของคนนั้น
license_id_of_person = person_with_plate['id'].values[0]

# หาข้อมูลส่วนตัวจากตาราง person โดยใช้ license_id
person_info = person[person['license_id'] == license_id_of_person]

# แสดงข้อมูลส่วนตัว
display(person_info)
```
