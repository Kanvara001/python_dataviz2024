# Quiz9
- วนลูปนับว่านิทานที่มีคำว่า แม่ หรือ พ่อ ในชื่อเรื่อง ทั้งหมดกี่เรื่อง เรื่องอะไรบ้าง
- มีคำดังกล่าวใน นิทานชีวิต หรือ นิทานประจำถิ่นมากกว่ากัน
- คนแต่งคนไหน ชอบแต่งนิทานที่มีชื่อเรื่องว่า แม่ หรือ พ่อ มากที่สุด (พร้อมทั้งอธิบายCode แต่ละบรรทัดด้วย)

````
count = 0
stories_with_parent = []
for index, row in this_data.iterrows():
    if 'แม่' in row['title'] or 'พ่อ' in row['title']:
        count += 1
        stories_with_parent.append(row['title'])

print(f"มีนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องทั้งหมด {count} เรื่อง")
print("รายชื่อนิทาน:")
for story in stories_with_parent:
    print(story)

count_ชีวิต = 0
count_ประจำถิ่น = 0

for index, row in this_data.iterrows():
    if 'แม่' in row['title'] or 'พ่อ' in row['title']:
        if row['type'] == 'นิทานชีวิต':
            count_ชีวิต += 1
        elif row['type'] == 'นิทานประจำถิ่น':
            count_ประจำถิ่น += 1

print(f"มีคำว่า 'แม่' หรือ 'พ่อ' ในนิทานชีวิต {count_ชีวิต} เรื่อง")
print(f"มีคำว่า 'แม่' หรือ 'พ่อ' ในนิทานประจำถิ่น {count_ประจำถิ่น} เรื่อง")

author_counts = {}
for index, row in this_data.iterrows():
    if 'แม่' in row['title'] or 'พ่อ' in row['title']:
        author = row['creator']
        if author in author_counts:
            author_counts[author] += 1
        else:
            author_counts[author] = 1

most_frequent_author = max(author_counts, key=author_counts.get)
print(f"คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่า 'แม่' หรือ 'พ่อ' มากที่สุดคือ: {most_frequent_author}")
````

บรรยายเกี่ยวกับผู้กำกับชั้นนำของอินเดีย 3 คน ว่าทำไมเราถึงคิดว่าเป็นผู้กำกับชั้นนำ และชั้นนำยังไง (พรีเซ้นคนละ3นาที)

````
import pandas as pd
````

````
import os
````

````
from google.colab import drive
drive.mount('/content/drive')
````

````
movie_df =  pd.read_csv('/content/drive/MyDrive/data_viz_2024_Data/IMDb Movies India.csv',encoding='latin-1')
print(movie_df.shape)
movie_df.head()
````

````
# หาจำนวนภาพยนตร์ที่แต่ละผู้กำกับได้กำกับ
director_counts = movie_df['Director'].value_counts()


#หาค่าเฉลี่ยของ Rating สำหรับแต่ละผู้กำกับ
director_avg_rating = movie_df.groupby('Director')['Rating'].mean().reset_index()


#เรียงลำดับผู้กำกับตามค่าเฉลี่ยของ Rating จากมากไปน้อย
sorted_directors = director_avg_rating.sort_values(by='Rating', ascending=False)
print("Top directors by average rating:")
for index, row in sorted_directors.head(3).iterrows():
    print(f"{row['Director']}: {row['Rating']:.2f}")

# Convert 'Votes' column to numeric, handling non-numeric values
#แปลงคอลัมน์ 'Votes' เป็นข้อมูลตัวเลข โดยจะจัดการกับค่าที่ไม่ใช่ตัวเลขได้ด้วยการกำหนด errors='coerce'
# ซึ่งมีความหมายคือ ถ้าพบค่าที่ไม่ใช่ตัวเลขจะแปลงให้เป็น NaN (Not a Number) แทน
movie_df['Votes'] = pd.to_numeric(movie_df['Votes'], errors='coerce')

#เรียงลำดับผู้กำกับตามจำนวนโหวต (Votes) จากมากไปน้อย
top_directors_by_votes = movie_df.groupby('Director')['Votes'].sum().sort_values(ascending=False).head(3)
print("Top directors by total votes:")
for director, total_votes in top_directors_by_votes.items():
    print(f"{director}: {total_votes}")
````
